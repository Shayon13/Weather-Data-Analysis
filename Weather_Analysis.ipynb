find all instance when Snow was recorded in weather condition

```python
print(df.head(1))
print(df['Weather condition'].value_counts())
# its only find the word snow 
snow_record=df[df['Weather condition']=='Snow']
print(snow_record)
#This show there is some other value in weather condition related to snow 
snow_record=df[df['Weather condition'].str.contains('Snow')]
print(snow_record)
```

find all instance when wind speed above 24 and visibility is 25
